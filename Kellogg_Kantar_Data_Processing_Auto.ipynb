```python
from openpyxl import load_workbook
import pandas as pd
import os

directory = r"C:\Users\KOCH1J02\OneDrive - Kellogg Company\Desktop\Kantar merged\52 weeks" 

files = os.listdir(directory)

Lheader = []
Ldata   = []
for file in files:
    if not file.endswith(".xlsx"):
        continue
    wb = load_workbook(os.path.join(directory, file))
    ws = wb[wb.sheetnames[0]]
    if not Lheader:
        Lheader = ["Date"] + [cell.value for cell in ws["B9:CX9"][0][1:]] + ["Measure", "Duration", "Outlet"]
    value = ws["B5"].value
    metadata = [value.split(" | ")[idx] for idx in [0, 2, -1]]
    
    for row in ws["B10:CX36"]:
        Ldata.append([cell.value for cell in row] + metadata)
        
df = pd.DataFrame(Ldata, columns=Lheader)

df.to_csv("52weeks.csv")
```
